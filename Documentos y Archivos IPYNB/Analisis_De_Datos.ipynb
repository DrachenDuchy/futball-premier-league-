<strong> Consulta </strong>

```
%sql

-- H= Victoria Local

-- A= Victoria Visitante

-- D= Empate

-- consulta de los partidos que estuvieron presentes, muestra el dia y fecha, los 2 equipos que jugaron (si fueron visitantes o locales) y si ganaron de locales


select HomeTeam, AwayTeam, HTR, Season, DateTime from data_results where HTR = 'H'
```

```
%sql

-- H= Victoria Local

-- A= Victoria Visitante

-- D= Empate

-- consulta de los partidos que estuvieron presentes, muestra el dia y fecha, los 2 equipos que jugaron (si fueron visitantes o locales) y si ganaron de visitantes

select HomeTeam, AwayTeam, HTR, Season, DateTime from data_results where HTR = 'A'
```
