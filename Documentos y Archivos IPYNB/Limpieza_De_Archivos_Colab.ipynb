## limpieza de datos nulos





Valores nulos

```python
cantidad_nulos1 = data1.isnull().sum()  # conteo de los nulos para players
print("Cantidad de campos vacios en cada columna: \n", cantidad_nulos1)
```

```python
cantidad_nulos2 = data2.isnull().sum()  # conteo de los nulos para results
print("Cantidad de campos vacios en cada columna: \n", cantidad_nulos2)
```

```python
# eliminar los datos vacios de las columnas de la base de datos Results

data2 = data2[~data2.Referee.isnull()].copy()
data2 = data2[~data2.HS.isnull()].copy()
data2 = data2[~data2.AS.isnull()].copy()
data2 = data2[~data2.HST.isnull()].copy()
data2 = data2[~data2.AST.isnull()].copy()
data2 = data2[~data2.HC.isnull()].copy()
data2 = data2[~data2.AC.isnull()].copy()
data2 = data2[~data2.HF.isnull()].copy()
data2 = data2[~data2.AF.isnull()].copy()
data2 = data2[~data2.HY.isnull()].copy()
data2 = data2[~data2.HR.isnull()].copy()
data2 = data2[~data2.AR.isnull()].copy()
```

```python
cantidad_nulos2 = data2.isnull().sum()  # rectificar que no halla valores vacios en results
print("Cantidad de campos vacios en cada columna: \n", cantidad_nulos2)
```
